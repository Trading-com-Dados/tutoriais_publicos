```python
import yfinance as yf
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
vale = yf.download('VALE3.SA', 
                  start = '2021-01-01', 
                  end = '2022-09-01')
```

```python
vale.head()
```

```python
vale['Adj Close'].plot();
```

```python
# Cria a figura
fig = make_subplots(rows = 1, cols = 1)

# Adiciona o gráfico de candlestick com os parâmetros OHLC no eixo y
# Esses parâmetros vão ser utilizados para criar os candles
# No eixo x adicionamos o index com as datas
fig.add_trace(go.Candlestick(name = 'Variação da cotação de VALE3',
                             x = vale.index, open = vale['Open'], high = vale['High'],
                             low = vale['Low'], close = vale['Close'],
                             showlegend = True), row = 1, col = 1)

# Personalizar o eixo Y
fig.update_yaxes(title_text = "<b> Preço VALE3 (R$)", row = 1, col = 1)

# Personalizar aspectos da figura em geral, como tamanho
fig.update_layout(xaxis_rangeslider_visible = True, width = 1000, height = 500)

# Mostra a figura
fig.show()
```
