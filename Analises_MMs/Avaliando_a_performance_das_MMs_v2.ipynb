# **2. Como verificar se o preço respeita a média móvel**

## **2.1. Visualizar as MMs como suporte/resistência**

Uma das melhores formas de definir se "o preço respeita a média móvel" seria avaliar se ela atua bem como um **suporte/resistência**

Dados OHLC MGLU3 pelo Yahoo Finance

```python
ativo_ohlc_1D = vbt.YFData.download('MGLU3.SA', start='2012-01-01', end= '2022-06-30').get(['Open','High','Low','Close'])
```

```python
ativo_ohlc_1D
```

Calcular duas Médias Móveis: 9, 20 períodos

```python
ativo_ohlc_1D['MMA_9p'] = vbt.MA.run(ativo_ohlc_1D.Close, 9, short_name='MMA_9p').ma
ativo_ohlc_1D['MMA_20p'] = vbt.MA.run(ativo_ohlc_1D.Close, 20, short_name='MMA_20p').ma
ativo_ohlc_1D
```

Note que, no período destacado no gráfico, o preço sempre se manteve acima da MM20p

```python
cotacoes_ativo_fig = ativo_ohlc_1D[(ativo_ohlc_1D.index > '2019-05-20') & (ativo_ohlc_1D.index <= '2019-08-28')]

fig = go.Figure(data=[go.Candlestick(go.Candlestick(name='MGLU3', x=cotacoes_ativo_fig.index, open=cotacoes_ativo_fig['Open'],high=cotacoes_ativo_fig['High'],
                                                    low=cotacoes_ativo_fig['Low'],close=cotacoes_ativo_fig['Close'],increasing_line_color= 'green', decreasing_line_color= 'red'))])

fig.add_trace(go.Scatter(name='MMA_9p', x=cotacoes_ativo_fig.index, y=cotacoes_ativo_fig['MMA_9p'], line=dict(color='red', width=3)))
fig.add_trace(go.Scatter(name='MMA_20p', x=cotacoes_ativo_fig.index, y=cotacoes_ativo_fig['MMA_20p'],marker_color='blue'))

fig.update_xaxes(title_text="<b> Data", rangebreaks=[dict(bounds=["sat", "mon"]), dict(values=["2017-12-30","2017-12-31","2017-12-31"])])

fig.add_vrect(x0='2019-05-28', x1='2019-08-21 12:00', fillcolor="green", opacity=0.25, line_width=0)

fig.update_traces(line_width=1)
fig.update_yaxes(title_text="<b>MGLU3 (R$)")
fig.update_layout(xaxis_rangeslider_visible=False,title_text='<b>Fechamento respeitou a MM20p',template = 'simple_white',margin=dict(l=20, r=20, t=20, b=20), paper_bgcolor="#f7f8fa",width=800,height=600,
                  font=dict(family="Arial",size=15,color="black"), legend=dict(orientation="h",yanchor="bottom",y=1.02,xanchor="right",x=1))
fig.show()
```

## **2.2. Calcular os retornos acumulados através do backtesting**

Exemplo com a MM aritmética de 50 períodos

```python
ativo_close_1D = vbt.YFData.download('MGLU3.SA', start='2012-01-01', end= '2022-06-30').get('Close')

MMA_50p = vbt.MA.run(ativo_close_1D, 50, short_name='MMA_50p')
entries = ativo_close_1D.vbt.crossed_above(MMA_50p.ma)
exits = ativo_close_1D.vbt.crossed_below(MMA_50p.ma)

pf_kwargs = dict(size=np.inf, freq='1D', fees=0.005, direction='both')
pf_MM50p = vbt.Portfolio.from_signals(ativo_close_1D, entries, exits, **pf_kwargs)
```

```python
pf_MM50p.plot(width = 500, height = 800, paper_bgcolor="#f7f8fa").show()
```

```python
pf_MM50p.total_return()
```

```python
pf_Buy_Hold = vbt.Portfolio.from_holding(ativo_close_1D)
pf_Buy_Hold.total_return()
```

## **3.3. Calcular as Médias Móveis aritméticas e exponenciais de 5 a 200 períodos**

```python
MMs_5_200p = list(range(5, 201))
```

```python
str(MMs_5_200p)
```

```python
MMA_5_200p_ativo = vbt.MA.run(ativo_close_1D, MMs_5_200p, short_name='MMA').ma
MME_5_200p_ativo = vbt.MA.run(ativo_close_1D, MMs_5_200p, short_name='MME', ewm=True).ma
```

```python
MME_5_200p_ativo.head()
```

```python
MME_5_200p_ativo.columns = MME_5_200p_ativo.columns.droplevel(1)
MME_5_200p_ativo
```

## **3.4. Qual MM trouxe melhor desempenho? Backtest Cruzam. Close vs MMs**

```python
entries_MMA = ativo_close_1D.vbt.crossed_above(MMA_5_200p_ativo)
exits_MMA = ativo_close_1D.vbt.crossed_below(MMA_5_200p_ativo)
entries_MME = ativo_close_1D.vbt.crossed_above(MME_5_200p_ativo)
exits_MME = ativo_close_1D.vbt.crossed_below(MME_5_200p_ativo)
entries_MME
```

Visualizar os retornos em relação aos diferentes períodos de média móvel utilizada no cruzamento

Considerando que temos algumas taxas durante nas operações (vamos supor 0,05% da transação), vejamos como ficariam os retornos

```python
pf_cruz_preco_MMA_taxas = vbt.Portfolio.from_signals(ativo_close_1D, entries_MMA, exits_MMA, fees=0.005, direction='both', freq='D')
pf_cruz_preco_MME_taxas = vbt.Portfolio.from_signals(ativo_close_1D, entries_MME, exits_MME, fees=0.005, direction='both', freq='D')
```

```python
pf_cruz_preco_MMA_taxas.total_return().vbt.plot(width = 800, height = 500, template = 'simple_white',paper_bgcolor="#f7f8fa").\
update_yaxes(title_text="<b> Total Return MGLU3 (com taxas)").update_xaxes(title_text="<b> MMA do Cruzamento").show()

pf_cruz_preco_MME_taxas.total_return().vbt.plot(width = 800, height = 500, template = 'simple_white',paper_bgcolor="#f7f8fa").\
update_yaxes(title_text="<b> Total Return MGLU3 (com taxas)").update_xaxes(title_text="<b> MME do Cruzamento").show()
```

Visualizando as entradas de das MM aritméticas de 50 períodos

```python
MMA_50p_ativo = vbt.MA.run(ativo_close_1D, 50, short_name='MMA_50p').ma
entries_50p = ativo_close_1D.vbt.crossed_above(MMA_50p_ativo)
exits_50p = ativo_close_1D.vbt.crossed_below(MMA_50p_ativo)
```

```python
pf_MM_50p = vbt.Portfolio.from_signals(ativo_close_1D, entries_50p, exits_50p, fees=0.005, direction='both')
fig = ativo_close_1D.vbt.plot(trace_kwargs=dict(name='Close'))
MMA_50p_ativo.vbt.plot(trace_kwargs=dict(name='MMA_50p'), fig=fig)
pf_MM_50p.positions.plot(close_trace_kwargs=dict(visible=False), fig=fig).show()
```

## **3.5. Calcular algumas métricas mais específicas do backtesting**

Visualizar os retornos das **MMS ARITMÉTICA X EXPONENCIAL** em relação aos diferentes períodos de média móvel utilizada no cruzamento

```python
pf_cruz_preco_MMA_taxas.total_return().vbt.plot(width=400, height=250, template='simple_white', paper_bgcolor="#f7f8fa").\
update_yaxes(title_text="<b>Total Return MGLU3").update_xaxes(title_text="<b>MMA do Cruzamento").show()

pf_cruz_preco_MME_taxas.total_return().vbt.plot(width=400, height=250, template='simple_white', paper_bgcolor="#f7f8fa",showlegend=False).\
update_yaxes(title_text="<b>Total Return MGLU3").update_xaxes(title_text="<b>MME do Cruzamento").show()
```

```python
pf_cruz_preco_MMA_taxas.trades.profit_factor().vbt.plot(width=400, height=250, template='simple_white', paper_bgcolor="#f7f8fa").\
update_yaxes(title_text="<b>Profit factor MGLU3").update_xaxes(title_text="<b>MMA do Cruzamento").show()

pf_cruz_preco_MME_taxas.trades.profit_factor().vbt.plot(width=400, height=250, template='simple_white', paper_bgcolor="#f7f8fa",showlegend=False).\
update_yaxes(title_text="<b>Profit factor MGLU3").update_xaxes(title_text="<b>MME do Cruzamento").show()
```

```python
pf_cruz_preco_MMA_taxas.drawdowns.max_drawdown().vbt.plot(width=400, height=250, template='simple_white', paper_bgcolor="#f7f8fa").\
update_yaxes(title_text="<b>Max drawdown MGLU3").update_xaxes(title_text="<b>MMA do Cruzamento").show()

pf_cruz_preco_MME_taxas.drawdowns.max_drawdown().vbt.plot(width=400, height=250, template='simple_white', paper_bgcolor="#f7f8fa",showlegend=False).\
update_yaxes(title_text="<b>Max drawdown MGLU3").update_xaxes(title_text="<b>MME do Cruzamento").show()
```
