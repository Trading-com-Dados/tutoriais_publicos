##Desemprego

```python
desemprego = wb.data.DataFrame('SL.UEM.TOTL.ZS', economy=['ARG', 'BRA', 'CHL', 'COL', 'PER', 'URY', 'MEX'], time=range(2010,2021))
desemprego
```

```python
#Lista de países

lista_paises = np.array([desemprego.index])
```

```python
#Funçao Melt

desemprego = desemprego.melt(var_name='Ano', value_name='Desemprego')
desemprego
```

```python
#Função resize

desemprego['País'] = np.resize(lista_paises, len(desemprego))
desemprego
```

```python
desemprego['Ano'] = pd.to_numeric(desemprego['Ano'].str.slice(start=2))
desemprego
```

```python
desemprego['Desemprego'] = pd.to_numeric(desemprego['Desemprego'])
desemprego
```

```python
#Gráfico de Barras Animado

fig = px.bar(desemprego, 
             x='País',
             y=['Desemprego'],
             color='País',
             animation_frame='Ano',
             animation_group='País',
             range_y=[0,14],
             title='Desemprego América Latina 2010-2020',
             template='plotly_dark')
fig.show()
```

##PIB Per Capita

```python
lista=['ARG', 'BRA', 'CHL', 'COL', 'PER', 'URY', 'MEX']
```

```python
pib_per_capita = wb.data.DataFrame('NY.GDP.PCAP.KD', economy=lista, time=range(2011,2021))
pib_per_capita = pib_per_capita.melt(var_name='Ano', value_name='pib_per_capita')
pib_per_capita['País'] = np.resize(lista_paises, len(pib_per_capita))
pib_per_capita['Ano'] = pd.to_numeric(pib_per_capita['Ano'].str.slice(start=2))
pib_per_capita['pib_per_capita'] = pd.to_numeric(pib_per_capita['pib_per_capita'])
pib_per_capita
```

```python
fig = px.scatter_geo(pib_per_capita,
                     locations='País',
                     color='País',
                     hover_name='País',
                     size=pib_per_capita['pib_per_capita'],
                     animation_frame='Ano',
                     projection='natural earth',
                     title='PIB per Capita America Latina',
                     template='simple_white')
fig.show()
```

##Formação Bruta de Capital Fixo

```python
fbcf = wb.data.DataFrame('NE.GDI.TOTL.ZS', economy=lista, time=range(2009,2021))
fbcf = fbcf.melt(var_name='Ano', value_name='fbcf')
fbcf['País'] = np.resize(lista_paises, len(fbcf))
fbcf['Ano'] = pd.to_numeric(fbcf['Ano'].str.slice(start=2))
fbcf['fbcf'] = pd.to_numeric(fbcf['fbcf'])
fbcf
```

```python
px.scatter(fbcf,
           x='País',
           y='fbcf',
           color='País',
           size='fbcf',
           animation_frame='Ano',
           animation_group='País',
           title='Formação Bruta de Capital Fixo',
           hover_name='País',
           range_y=[10,35])
```
