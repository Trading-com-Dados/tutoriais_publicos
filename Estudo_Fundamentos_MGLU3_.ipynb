##3. Coletar os preços de MGLU 

```python
#Coletando Preços

mglu = investpy.get_stock_historical_data('MGLU3', country='Brazil', from_date='01/01/2018', to_date='08/12/2021')
```

```python
#Gráfico que responde a pergunta: como os preços das ações estão respondendo ao IPCA

#Comportamento preços vs IPCA

from plotly.subplots import make_subplots

fig = make_subplots(specs=[[{"secondary_y": True}]])
fig.add_trace(go.Scatter(x=ipca.index, y=ipca['Value'], name='IPCA'), secondary_y=False)
fig.add_trace(go.Scatter(x=mglu.index, y=mglu['Close'], name='MagaluPreço'),secondary_y=True)
fig.update_layout(title_text='Preço vs IPCA')
fig.show()
```

```python
#QUal o nível de incerteza no mercado?

#Janela de Voltilidade 30 dias - alta indefinição por parte do mercado
mglu_retornos = mglu['Close'].pct_change()
rol_vol = mglu_retornos.rolling(30).std()

fig = go.Figure()
fig.add_trace(go.Scatter(x=rol_vol.index, y=rol_vol))
fig.update_layout(title_text= 'Janela Volatilidade 30 dias')
fig.show()
```

###4.1. Receitas e Margens

```python
indicadores = pd.read_excel(io.BytesIO(uploaded['RESULTADO_3T21_POR.xlsx']), header=1)
indicadores.set_index(indicadores['R$ milhões (exceto quando indicado)'],inplace=True)

indicadores
```

```python
indicadores.drop(columns=['R$ milhões (exceto quando indicado)'],inplace=True)
indicadores.drop(columns=['Unnamed: 24',	'2016',	'2017',	'2018',	'2019',	'2020',	'9M20',	'9M21'], axis=1, inplace=True)
indicadores
```

```python
#Ebitda vs Margem Ebtida, aqui temos a conclusão se a margem EBTIDA é ebtida/receita
#tenho que a empresa está sendo menos eficiente operacionalmente e custos, mais necessariamente
#despesas estão influenciando na queda da margem.

#Avaliando o EBTIDA
#O que é o EBTIDA - Earn before taxes interest depreciation and amortization, representa a receita referente
#as operações da emrpesa, geração de caixa operacional da companhia
#Ebtida vs Margem Ebtida

fig = make_subplots(specs=[[{"secondary_y": True}]])
fig.add_trace(go.Bar(x=indicadores.columns, y=indicadores.loc['EBITDA'],name='Ebtida', text=indicadores.loc['EBITDA']), secondary_y=False)
fig.add_trace(go.Scatter(x=indicadores.columns, y=indicadores.loc['Margem EBITDA'],name='Margem Ebtida'), secondary_y=True)
fig.update_layout(title_text='EBTIDA vs Margem Ebtida')
fig.show()
```

```python
#Como se comportam Margem Líquida e Margem Ebtida, aqui temos o resultado de
#receitas tributárias que foram contabilizadas no resultado e melhoraram a margem líquida
#Margem líquida vs Margem Ebtida

fig = go.Figure()
fig.add_trace(go.Scatter(x=indicadores.columns, y=indicadores.loc['Margem EBITDA'],name='Margem Ebtida'))
fig.add_trace(go.Scatter(x=indicadores.columns, y=indicadores.loc['Margem Líquida'], name='Margem Líquida'))
fig.update_layout(title_text='Margem Ebitda vs Margem Liquida')
fig.show()
```
