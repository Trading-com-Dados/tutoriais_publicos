## 2. Importando dados

Repare que vou importar também os dados da cotação do dólar. Posteriormente vou converter a cotação do Ethereum para reais, por isso que estou trazendo a cotação do dólar.

```python
ativos = ['PETR4.SA','VALE3.SA', 'WEGE3.SA','BPAC11.SA', 
          'RADL3.SA', 'OIBR3.SA','KNRI11.SA','BOVA11.SA',
          'SMAL11.SA','AAPL34.SA','IVVB11.SA']
```

```python
inicio = '2020-06-01'
fim = '2022-03-06'
```

```python
precos = yf.download(ativos, 
                     start = inicio, 
                     end = fim)['Adj Close']
```

Vamos normalizar o preço dos ativos para visualizar seus desempenhos

```python
precos.rename(columns = {'PETR4.SA':'PETR4','VALE3.SA':'VALE3', 
                         'WEGE3.SA':'WEGE3','BPAC11.SA':'BPAC11', 
                         'RADL3.SA':'RADL3', 'OIBR3.SA':'OIBR3',
                         'KNRI11.SA':'KNRI11','BOVA11.SA':'BOVA11',
                         'SMAL11.SA':'SMAL11','AAPL34.SA':'AAPL34',
                         'IVVB11.SA':'IVVB11'}, inplace = True)
```

```python
precos.plot(figsize = (10,10));
```

```python
normalizado = precos/precos.iloc[0]
```

```python
normalizado.plot(figsize = (10,10));
```
