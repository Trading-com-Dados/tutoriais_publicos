##Exercício

Criar um gráfico que sumarize um grupo de métricas fundamentalistas de um determinado setor

```python
#lista de empresas através do código do setor

fundamentus.list_papel_setor(8)
```

```python
setor = fundamentus.list_papel_setor(8)
setor
```

```python
df.query('index in @setor')
```

```python
df.query('index in @setor').loc[:,['mrgebit', 'mrgliq', 'roic', 'roe']]
```

```python
varejo = df.query('index in @setor').loc[:,['mrgebit', 'mrgliq', 'roic', 'roe']]
varejo
```

```python
fig = make_subplots(rows=2,
                    cols=2,
                    row_heights=[4,4],
                    column_widths=[4,4],
                    subplot_titles=('Margem EBTIDA', 'Margem Líquida', 'ROIC', 'ROE'),
                    shared_xaxes=False)
fig.add_trace(go.Bar(name='Margem_Ebtida', x=varejo.index, y=varejo['mrgebit']), row=1, col=1)
fig.add_trace(go.Bar(name='Margem_Liquida', x=varejo.index, y=varejo['mrgliq']), row=1, col=2)
fig.add_trace(go.Bar(name='ROIC', x=varejo.index, y=varejo['roic']), row=2, col=1)
fig.add_trace(go.Bar(name='ROE', x=varejo.index, y=varejo['roe']), row=2, col=2)
fig.update_layout(title_text='<b>Overview Rentabilidade Varejo<b>', 
                  template='plotly_dark',#template pré-definido da plotly
                  showlegend=False, #esconder ou mostrar legenda
                  height=500, #altura
                  width=900) #largura
fig.show()
```

```python
varejo2 = df.query('index in @setor').loc[:,['pl', 'pvp', 'evebitda', 'dy']]
varejo2
```

```python
#ORDENANDO OS VALORES

#Para fazer isso deixar o shared xaxes = False  e adicionar o fig.update_xaxes(categoryorder='total descending')

fig = make_subplots(rows=2,
                    cols=2,
                    row_heights=[5,5],
                    column_widths=[3,3],
                    subplot_titles=('Preço/Lucro', 'P/VP', 'EV/EBTIDA', 'Dividend Yeld'),
                    shared_xaxes=False)
fig.add_trace(go.Bar(name='Margem_Ebtida', x=varejo2.index, y=varejo2['pl']), row=1, col=1)
fig.add_trace(go.Bar(name='Margem_Liquida', x=varejo2.index, y=varejo2['pvp']), row=1, col=2)
fig.add_trace(go.Bar(name='ROIC', x=varejo2.index, y=varejo2['evebitda']), row=2, col=1)
fig.add_trace(go.Bar(name='ROE', x=varejo2.index, y=varejo2['dy']), row=2, col=2)
fig.update_layout(title_text='<b>Avaliação por Múltiplos Varejo<b>', 
                  template='plotly_dark',#template pré-definido da plotly
                  showlegend=False, #esconder ou mostrar legenda
                  height=500, #altura
                  width=900) #largura
fig.update_xaxes(categoryorder='total descending')
fig.show()
```
