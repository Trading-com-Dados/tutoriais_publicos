### Importação

```python
import pandas as pd
import yfinance as yf
from ta import add_all_ta_features
import pandas_ta
import btalib
import talib
from finta import TA
```

```python
data = yf.download('LREN3.SA', start = '2020-01-01')
```

## **Biblioteca TA**

```python
data = add_all_ta_features(
    data, open="Open", high="High", low="Low", close="Close", volume="Volume")
```

```python
data.trend_macd.plot()
```

## **Biblioteca BTA-LIB**

```python
macd = btalib.macd(data.Close)
macd.df.plot()
```

## **Biblioteca pandas_ta**

```python
data.ta.macd().plot()
```

## **Biblioteca ta-lib**

```python
output = talib.MACD(data.Close)
```

```python
(pd.DataFrame(output).T).plot()
```
